```python
import datetime
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras import Sequential, Input
from tensorflow.keras.layers import Dense, Embedding, LSTM, TimeDistributed, Dropout, AlphaDropout, BatchNormalization, LeakyReLU
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.optimizers import *
```

```python
f_data = open('emnist/emnist-byclass-'+ 'train' + '-images-idx3-ubyte')
```

```python
def load_data(type_text):
    f_data = open('emnist/emnist-byclass-'+ type_text + '-images-idx3-ubyte')
    f_data = np.fromfile(file=f_data, dtype=np.uint8)
    f_data = f_data[16:].reshape(-1, 28, 28).astype(np.uint8)
    f_label = open('emnist/emnist-byclass-'+ type_text + '-labels-idx1-ubyte')
    f_label = np.fromfile(file=f_label, dtype=np.uint8)
    f_label = f_label[8:].reshape(-1).astype(np.uint8)
    return f_data, f_label
```

```python
train_data, train_label = load_data('train')
test_data, test_label = load_data('test')
```

```python
# Validation data size: Train=697,932, Test=116,323
print("Train data size: {train}, Test data size: {test}".format(train=len(train_data), test=len(test_data)))
```

```python
train_label = to_categorical(train_label, num_classes=62, dtype='uint8')
test_label = to_categorical(test_label, num_classes=62, dtype='uint8')
```

```python
train_data = train_data.reshape(-1, 784)
test_data = test_data.reshape(-1, 784)
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# create model
model = Sequential()

# =====STABLE=====
model.add(Dense(500, input_shape=(784,), activation='swish'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(384, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(192, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(62, activation='softmax'))
# ================

# # Compile model
opt = Adam(learning_rate=0.005)
model.compile(loss='CategoricalCrossentropy', optimizer=opt, metrics=['accuracy'])
plot_model(model, to_file="model.png", show_shapes=True)
```

```python
# Baseline: Test Data Accuracy = 0.857
model.fit(train_data, train_label, epochs=20, batch_size=128, verbose=2, shuffle=True, validation_split=0.1, callbacks=[tensorboard_callback])
```

```python
# create model
model_2 = Sequential()

# =====STABLE=====
model_2.add(Dense(512, input_shape=(784,), activation='swish'))
model_2.add(AlphaDropout(0.3))
model_2.add(Dense(256, activation='relu'))
model_2.add(AlphaDropout(0.3))
model_2.add(Dense(128, activation='relu'))
model_2.add(AlphaDropout(0.3))
model_2.add(Dense(62, activation='softmax'))
# ================

# # Compile model
opt = Adam(learning_rate=0.005)
model_2.compile(loss='CategoricalCrossentropy', optimizer=opt, metrics=['accuracy'])
model_2.summary()
```

```python
model_2.fit(train_data, train_label, epochs=20, batch_size=256, verbose=2, shuffle=True, validation_split=0.1, callbacks=[tensorboard_callback])
```

```python
# model.save('saved_model/1-1_model')
```

```python
results = model.evaluate(test_data, test_label)
```

```python
pred_Y = model.predict(test_data)
```

```python
# DRAW confusion-matrix
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pred_Y = np.argmax(pred_Y, axis=1)
```

```python
true_Y = np.argmax(test_label, axis=1)
```

```python
conf_matrix = confusion_matrix(true_Y, pred_Y)
```

```python
fig, ax = plt.subplots(figsize=(60,60))         # Sample figsize in inches
fig = sns.heatmap(conf_matrix, annot=True, fmt='d', linewidths=.5, ax=ax)
```
